```python
def remove_substring(name):
    start = name.find("(")
    end = name.find(")")
    new_name = name[0:start] + name[end+1:]
    new_name = new_name.strip()
    return new_name
```

```python
df["Name_trimmed"] =df['Name'].apply(remove_substring)
```

```python
df["Name_trimmed"].head()
```

## Replace substring with Regex : regular expression 

```python
# Replace substring with regex : regular expression
#   * means zero or more
#   . match any charactor 
# "\(.*\)"

# in windows we use \\  double slash
df['Name'].str.replace(r"\(.*\)","HELLO", regex=True).head()
```

```python
# in window if we use r strip then with single slash also it works
df['Name'].str.replace("\\(.*\\)","HELLO", regex=True).head()
```

# Categorical Data in Pandas 

## What is Categorical Data ?

### Categorical Data is textual or labeled data that represents discrete cotegories 

```python
class_data = {
      "Name":["A","B","C","D"],
     "Groups":["G1","G2","G3","G4"]  
}
```

```python
class_df = pd.DataFrame(class_data)
```

```python
class_df.head()
```

```python
class_df.info()
```

```python
class_df['Group_cat'] = pd.Categorical(
    class_df['Groups'],
    categories =['G3','G1','G2','G4'], # custom order Higher to lower 
    ordered = True
)
```

```python
class_df.info()
```

```python
class_df['Group_cat'].cat.categories
```
